#畫函數圖形
畫 $f(x)=\sin(2x)+\cos(x)+2\sin(x)$


```python
# prompt: 畫出y=sin(x)圖形

import numpy as np
import matplotlib.pyplot as plt

# 定義函數
def f(x):
  return np.sin(2*x) + np.cos(x) + 2*np.sin(x)

# 建立 x 座標的數值範圍
x = np.linspace(-2*np.pi, 2*np.pi, 400)

# 計算對應的 y 座標
y = f(x)

# 畫出圖形
plt.plot(x, y)

# 設定圖表標題和座標軸標籤
plt.title('$f(x) = \sin(2x) + \cos(x) + 2\sin(x)$')
plt.xlabel('x')
plt.ylabel('y')

# 顯示網格線
plt.grid(True)

# 顯示圖形
plt.show()
```
